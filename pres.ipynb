```python
import pytz
from datetime import datetime

def get_canadian_timezones():
  """Returns a list of common Canadian time zones."""
  return [
      'America/Vancouver',
      'America/Regina',
      'America/Edmonton',
      'America/Winnipeg',
      'America/Toronto',
      'America/Halifax',
      'America/St_Johns',
      'America/Moncton'
  ]

def get_times_for_timezones(est_datetime):
  """Returns a dictionary of times for different Canadian time zones.

  Args:
    est_datetime: A datetime object in EST timezone.

  Returns:
    A dictionary where keys are time zone names and values are datetime objects.
  """

  est_timezone = pytz.timezone('US/Eastern')
  est_datetime = est_timezone.localize(est_datetime)

  timezones = get_canadian_timezones()
  time_dict = {}
  for timezone in timezones:
    target_timezone = pytz.timezone(timezone)
    target_datetime = est_datetime.astimezone(target_timezone)
    time_dict[timezone] = target_datetime
  return time_dict

# Example usage:
est_time = datetime(2024, 8, 12, 17, 0) #datetime.now() #datetime(2024, 8, 12, 12, 0)  # Replace with your EST datetime
times = get_times_for_timezones(est_time)
for timezone, time in times.items():
  print(f"{timezone}: {time} {(time.strftime('%Z'))}")
# for timezone, (time, code) in times.items():
#   print(f"{timezone}: {time} ({code})")
```
